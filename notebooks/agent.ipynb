```python
from trader.dao.dao import DAO

dao = DAO()
```

```python
import matplotlib.pyplot as plt
from datetime import datetime

from trader.dao.agent_histories import get_agent_histories

x = []
y = []
agent_histories_data = get_agent_histories(engine=dao.engine, limit=100000)
for agent_history in agent_histories_data:
    x.append(agent_history.created_at)
    y.append(agent_history.credits)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.yaxis.get_major_formatter().set_scientific(False)
ax.yaxis.get_major_formatter().set_useOffset(False)
fig.autofmt_xdate()
```
